Combine orders and customers tables to get more detailed information about each order

```python
%%sql
SELECT  o.order_id,
        c.customer_id,
        c.company_name,
        c.contact_name,
        c.city ||', '|| c.country AS location,
        o.ship_address,
        o.order_date,
        o.shipped_date
  FROM customers c
  JOIN orders o
    ON o.customer_id = c.customer_id
 LIMIT 10;
```

Combine employees and orders tables to see who is responsible for each order

```python
%%sql
SELECT e.employee_id,
       o.order_id,
       e.first_name ||' '|| e.last_name AS employee_name,
       e.title,
       o.order_date   
  FROM employees e
  JOIN orders o
    ON e.employee_id = o.employee_id
 ORDER BY o.order_date DESC
 LIMIT 10;
```

Combine order_details, products, and orders tables to get detailed order information, including the product name and quantity.

```python
%%sql
SELECT o.order_id,
       p.product_id,
       p.product_name,
       od.quantity,
       o.order_date
  FROM order_details od
  JOIN products p
    ON od.product_id = p.product_id
  JOIN orders o
    ON od.order_id = o.order_id
 LIMIT 10;
```

Ranking Employees by Sale Performance

```python
%%sql --Total sales for each employee + Ranking
WITH
employees_sales AS (
SELECT o.order_id,
       od.unit_price,
       od.quantity,
       o.customer_id,
       o.employee_id,
       e.first_name ||' '|| e.last_name AS name,
       od.discount,
       ROUND(SUM((od.unit_price * od.quantity) * (1 - od.discount)) OVER(PARTITION BY 
                                                                         o.employee_id)) AS total_sales_by_employee
                                            
  FROM orders o
  JOIN employees e
    ON o.employee_id = e.employee_id
  JOIN order_details od
    ON od.order_id = o.order_id
)
SELECT  es.name,
        es.employee_id,
        es.total_sales_by_employee,
        RANK () OVER(ORDER BY es.total_sales_by_employee DESC) AS sales_rank
  FROM employees_sales es
 GROUP BY es.name, es.employee_id, es.total_sales_by_employee
  


```

Margaret leads the team in sales while Steven is the lowest-selling employee on the team

Running Total of Sales per Month

```python
%%sql
WITH
monthly_sales AS(
SELECT o.order_id,
       DATE_TRUNC('MONTH', o.order_date)::DATE AS month,
       od.*,
       SUM((od.unit_price * od.quantity) * (1 - od.discount)) OVER(PARTITION BY DATE_TRUNC
                                                                          ('MONTH', o.order_date))
                                                                          AS sales_per_month
 FROM order_details od
 JOIN orders o 
   ON od.order_id = o.order_id
)
SELECT month,
       ROUND(sales_per_month::numeric, 2) AS sales_per_month_rounded,
       ROUND(SUM(sales_per_month) OVER (ORDER BY month)::numeric, 2) AS running_total_rounded
  FROM monthly_sales 
 GROUP BY month,sales_per_month
 ORDER BY month ASC
```

* The data suggests potential seasonal trends. For example, sales in December 1996 and 1997 are relatively high, which could indicate increased demand during the holiday season.
* Additionally, the sales in early months of the year (like January and February) show a tendency to be lower than in other months, potentially due to post-holiday season slowdown


MoM Sales Growth Rate(%)

```python
%%sql 
WITH
monthly_sales AS(
SELECT EXTRACT ('YEAR' FROM o.order_date) AS year,
       EXTRACT ('MONTH'FROM o.order_date) AS month,
       SUM((od.unit_price * od.quantity) * (1-od.discount)) month_sales
  FROM orders o
  JOIN order_details od
    ON o.order_id = od.order_id
 GROUP BY month, year

),
lagged_sales AS(
SELECT year,month,month_sales,
       LAG(month_sales,1,NULL) OVER (ORDER BY year, month) AS prev_month_sales
  FROM monthly_sales
)
SELECT year,
       month,
       ROUND(month_sales::numeric,2) AS "Month Sales",
       ROUND(prev_month_sales::numeric,2) AS "Previous Monthly Sales",
       ROUND(((month_sales - prev_month_sales)*100 / prev_month_sales)::NUMERIC,2) AS "Growth Rate"
  FROM lagged_sales
  


```

Seasonal Periods: 
There appears to be a pattern of sales peaking towards the end of the year, particularly in November and December of 1996 and 1997, followed by a sharp decline in February. This could indicate a strong seasonal influence, likely related to holiday shopping seasons.

Consistent Negative Growth Months:
Year 1996: August and December were the only months with negative growth.
Year 1997: February, June, August, and November showed negative growth rates.
Year 1998: May experienced a significant drop, the only month listed in 1998. 

Mean & Median Negative Growth Rates: The mean negative growth rate across the months that experienced a decline in sales compared to their prior month is approximately -29.45%. Meanwhile, the median is -32.39%. Negative Growth rates are right-skewed,likely significantly affected by -85.19% growth in May of 1998.

Identify Customers With Above-Average Order Values

```python
%%sql
WITH 
customer_order_value AS(
SELECT c.customer_id,
       c.contact_name,
       SUM((od.unit_price * od.quantity) * (1 - od.discount)) "Order Value"
  FROM order_details od
  JOIN orders o
    ON od.order_id = o.order_id
  JOIN customers c
    ON o.customer_id = c.customer_id
 GROUP BY c.customer_id, c.contact_name
)
SELECT cov.customer_id,
       cov.contact_name,
       "Order Value",
       CASE
           WHEN "Order Value" > AVG("Order Value") OVER () THEN 'Above Average'
           ELSE 'Below Average'
       END AS "Value Category"
  FROM customer_order_value cov
 ORDER BY "Order Value" DESC
 LIMIT 100;



 
 
```

Percentage of Total Sales for Each Product Category

```python
%%sql
WITH
total_sales AS(
SELECT c.category_id,
       c.category_name,
       SUM((p.unit_price * od.quantity) * (1 - od.discount)) "Total Sales"
       
  FROM categories c
  JOIN products p
    ON c.category_id = p.category_id
  JOIN order_details od
    ON p.product_id = od.product_id
 GROUP BY c.category_id, c.category_name
)
SELECT ts.category_id,
       ts.category_name,
       "Total Sales",
       "Total Sales" / SUM("Total Sales") OVER () *100.0 AS "Sales Percentage"
  FROM total_sales ts
 ORDER BY "Sales Percentage" DESC
```

The Beverage category leads the company sales as it composes 21.33% of total sales. Following shortly behind are Dairy Products at 18.56% and Confections at 13.29% of total sales.

Top 3 Products Sold In each Category

```python
%%sql
WITH 
product_sales AS(
SELECT c.category_id,
       p.product_id,
       p.product_name,
       SUM((p.unit_price * od.quantity) * (1 - od.discount)) "Total Sales"
  FROM products p
  JOIN order_details od
    ON p.product_id = od.product_id
  JOIN categories c
    ON p.category_id = c.category_id
 GROUP BY p.product_id, c.category_id, p.product_name
),
rownum AS(
SELECT ps.category_id,
       ps.product_id,
       ps.product_name,
       "Total Sales",
       ROW_NUMBER () OVER (PARTITION BY ps.category_id
                           ORDER BY "Total Sales" DESC)
                           AS "Product Category Rank"
  FROM product_sales ps
)
SELECT r.category_id, r.product_id, r.product_name,"Total Sales", "Product Category Rank"
  FROM rownum r
 WHERE "Product Category Rank" <4


```
